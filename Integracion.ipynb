# INTEGRACIÓN

En esta etapa importamos el resultado de las otras transformaciones siendo INUMET_INE_promedio.csv, UTE_INE_promedio, y ECH_promedio. Una vez cargados procedemos a ejecutar dos joins, primero UTE_INE_promedio con ECH_promedio y luego al resultado de este join se le ejecuta otro join con INUMET_INE_promedio. Finalmente se materializa dicha integracion en el archivo "Integracion.csv".

```python
import pandas as pd


InumetDF = pd.read_csv('INUMET/INUMET_INE.csv').rename({'seccion':'secc', 'segmento':'segm', 'Año':'anio', 'Mes':'mes'}, axis=1)

UteDF = pd.read_csv('UTE/UTE_INE_promedio.csv')

EchDF = pd.read_csv('ECH/ECH_promedio.csv').rename({'seccion':'secc', 'segmento':'segm', 'año':'anio'}, axis=1)

UteAndECH = UteDF.merge(EchDF, how='inner', on=['segm','secc', 'anio'])

UteAndECHAndInumet = UteAndECH.merge(InumetDF, how='inner', on=['anio', 'mes', 'secc', 'segm'])

UteAndECHAndInumet.to_csv('ECH_UTE_INUMET.csv',index=False)
```
